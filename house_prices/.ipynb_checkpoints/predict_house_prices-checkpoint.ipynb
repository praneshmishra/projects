#### Rank the categorical variables based on ANOVA

```python
#calculate disparity score, identify important categorical variable based on ANOVA

def anova(frame):
    anv = pd.DataFrame()
    anv['features'] = frame.columns
    pvals = []
    for c in frame.columns.values.tolist():
        samples = []
        for cls in frame[c].unique():
            s = frame[frame[c] == cls]['SalePrice'].values
            samples.append(s)
        pval = stats.f_oneway(*samples)[1]
        pvals.append(pval)            
    anv['pval'] = pvals
    return anv.sort_values('pval')

tmp_data= cat_data.copy()
tmp_data['SalePrice'] = num_data['SalePrice']
imp_features = anova(tmp_data)
imp_features['disparity'] = np.log(1/imp_features['pval'].values)
plt.figure(figsize=(20,6))
sns.barplot(data=imp_features, x = 'features', y='disparity')
plt.title('Important categorical variables based on ANOVA', fontsize=15)
plt.ylabel('Disparity', fontsize=15)
plt.xlabel('Features', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

### function for priting out the metrics

```python
#function for priting out the metrics
def print_metrics(model, train_test, y, y_pred, features):
    
    rsquared_adj = 'NA'
    print("***************************** START *******************************")
    print("Metrics output for - ", train_test, " Data")
    print("Model Used :\n ", model)
    RMSE = np.sqrt(np.mean(np.square(y - y_pred)))
    MAPE  = np.mean(np.abs(y - y_pred)/y)
    
    if type(model).__name__ == 'RegressionResultsWrapper':
        rsquared = model.rsquared
    else:
        rsquared = r2_score(y,y_pred)
    
    print('\n')
    print("rsquared : ",rsquared)
    print("rsquared_adj: ",rsquared_adj)
    print("Log RMSE : " , RMSE)
    print("MAPE : " , MAPE)
    print("***************************** END *******************************")
    
    df = pd.DataFrame(columns=metrics_df.columns)
    model_string = str(model).split('(')[0]
    df.loc[-1] = [model,model_string,train_test,MAPE,RMSE,rsquared,rsquared_adj,features]
    df.index = df.index + 1
    
    return df
```

```python
metrics_df = pd.DataFrame(columns=["Model","Model_Name","Train_Test","MAPE","RMSE","R2", "Adj. R2", "Features_used"])
```
