### WALMART STORE SALES PREDICTION - FORECASTING

#### objective:
The objective is predicting store sales using historical markdown data.
One challenge of modelling retail data is the need to make decisions based on limited history. If Christmas comes but once a year, so does the chance to see how strategic decisions impacted the bottom line.

The four holidays in the dataset fall within the following weeks:
    
1. Super Bowl: 12-Feb-10, 11-Feb-11, 10-Feb-12, 8-Feb-13
2. Labor Day: 10-Sep-10, 9-Sep-11, 7-Sep-12, 6-Sep-13
3. Thanksgiving: 26-Nov-10, 25-Nov-11, 23-Nov-12, 29-Nov-13
4. Christmas: 31-Dec-10, 30-Dec-11, 28-Dec-12, 27-Dec-13
    
We would create 4 columns for 4 holidays and then can drop IsHoliday column.


```python
##We would create 4 columns for 4 holidays and then can drop IsHoliday column - train data

train_data['SuperBowl']= np.where(train_data.Date=='2010-02-12',1,
                               np.where(train_data.Date=='2011-02-11',1,
                                        np.where(train_data.Date=='2012-02-10',1,
                                                 np.where(train_data.Date=='2013-02-08',1,0))))

train_data['LaborDay']= np.where(train_data.Date=='2010-09-10',1,
                               np.where(train_data.Date=='2011-09-09',1,
                                        np.where(train_data.Date=='2012-09-07',1,
                                                 np.where(train_data.Date=='2013-07-06',1,0))))

train_data['Thanksgiving']= np.where(train_data.Date=='2010-11-26',1,
                               np.where(train_data.Date=='2011-11-25',1,
                                        np.where(train_data.Date=='2012-11-23',1,
                                                 np.where(train_data.Date=='2013-11-29',1,0))))

train_data['Christmas']= np.where(train_data.Date=='2010-12-31',1,
                               np.where(train_data.Date=='2011-12-30',1,
                                        np.where(train_data.Date=='2012-12-28',1,
                                                 np.where(train_data.Date=='2013-12-27',1,0))))

train_data.drop(['IsHoliday'],axis=1,inplace=True)

train_data.head(5)
```

```python
##We would create 4 columns for 4 holidays and then can drop IsHoliday column - test data

test_data['SuperBowl']= np.where(test_data.Date=='2010-02-12',1,
                               np.where(test_data.Date=='2011-02-11',1,
                                        np.where(test_data.Date=='2012-02-10',1,
                                                 np.where(test_data.Date=='2013-02-08',1,0))))

test_data['LaborDay']= np.where(test_data.Date=='2010-09-10',1,
                               np.where(test_data.Date=='2011-09-09',1,
                                        np.where(test_data.Date=='2012-09-07',1,
                                                 np.where(test_data.Date=='2013-07-06',1,0))))

test_data['Thanksgiving']= np.where(test_data.Date=='2010-11-26',1,
                               np.where(test_data.Date=='2011-11-25',1,
                                        np.where(test_data.Date=='2012-11-23',1,
                                                 np.where(test_data.Date=='2013-11-29',1,0))))

test_data['Christmas']= np.where(test_data.Date=='2010-12-31',1,
                               np.where(test_data.Date=='2011-12-30',1,
                                        np.where(test_data.Date=='2012-12-28',1,
                                                 np.where(test_data.Date=='2013-12-27',1,0))))

test_data.drop(['IsHoliday'],axis=1,inplace=True)

test_data.head(5)
```

```python
#Changing Data type of Date Variable and Extracting month,Year,week and quater from it - train data

train_data['Date'] = pd.to_datetime(train_data.Date,format="%Y-%m-%d")
train_data['Month']= train_data.Date.dt.month
train_data['Year'] = train_data.Date.dt.year
train_data['Week'] = train_data.Date.dt.week
train_data['Quarter'] = train_data.Date.dt.quarter

train_data.drop(['Date'],axis=1,inplace=True)

train_data.head(10)
```

### Model building and verification

#### RandomForestRegressor with gridsearchCV

```python
#parameter grid for gridsearchCV - RandomForestRegressor

param_grid_rf = {'n_estimators': [20,30,40,50],
                  'max_depth': [5,6,7,8,9,10]}

gscv_rf = GridSearchCV(estimator=RandomForestRegressor(), 
                        param_grid=param_grid_rf, 
                        cv=5,
                        verbose=True, n_jobs=-1)

gscv_results = gscv_rf.fit(train_x, train_y)
```

```python
gscv_results.best_params_
```

```python
radm_clf = gscv_rf.best_estimator_

indices = np.argsort(radm_clf.feature_importances_)[::-1]
feature_rank = pd.DataFrame( columns = ['rank', 'feature', 'importance'] )
for f in range(train_x.shape[1]):
  feature_rank.loc[f] = [f+1,
                         train_x.columns[indices[f]],
                         radm_clf.feature_importances_[indices[f]]]
plt.figure(figsize=(12,6))
sns.barplot( y = 'feature', x = 'importance', data = feature_rank )
plt.grid()
plt.title('Important Variables - Random Forest ranking', fontsize=12)
plt.ylabel('Variables', fontsize=12)
plt.xlabel('Importance', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

#### Decile analysis

```python
#Decile analysis - Train
train_y_pred['pred'] = train_pred 

train_y_pred['Deciles']=pd.qcut(train_y_pred['pred'],10, labels=False)

avg_actual = train_y_pred[['Deciles','Weekly_Sales']].groupby(train_y_pred.Deciles).mean().sort_index(ascending=False)['Weekly_Sales']
avg_pred = train_y_pred[['Deciles','pred']].groupby(train_y_pred.Deciles).mean().sort_index(ascending=False)['pred']

Decile_analysis_train = pd.concat([avg_actual, avg_pred], axis=1)

Decile_analysis_train
```

```python
plt.figure(figsize=(12,6))
#sns.set_style("ticks", {"xtick.major.size":2,"ytick.major.size":2})
sns.lineplot(Decile_analysis_train.index, Decile_analysis_train.Weekly_Sales,markers=True)
sns.lineplot(Decile_analysis_train.index, Decile_analysis_train.pred,markers=True)
plt.legend(['Actual','Predicted'], loc='best', fontsize=16)
plt.title('Decile Analysis - Total Weekly Sales / Predicted - Train Data', fontsize=18)
plt.ylabel('Total Weekly Sales', fontsize=16)
plt.xlabel('Deciles', fontsize=16)
plt.show()
```

```python
#Decile analysis - Test
test_y_pred['pred'] = test_pred 
test_y_pred['Deciles']=pd.qcut(test_y_pred['pred'],10, labels=False)

avg_actual_test = test_y_pred[['Deciles','Weekly_Sales']].groupby(test_y_pred['Deciles']).mean().sort_index(ascending=False)['Weekly_Sales']
avg_pred_test = test_y_pred[['Deciles','pred']].groupby(test_y_pred.Deciles).mean().sort_index(ascending=False)['pred']

Decile_analysis_test = pd.concat([avg_actual_test, avg_pred_test], axis=1)

Decile_analysis_test
```

```python
plt.figure(figsize=(12,6))
#sns.set_style("ticks", {"xtick.major.size":2,"ytick.major.size":2})
sns.lineplot(Decile_analysis_test.index, Decile_analysis_test.Weekly_Sales,markers=True)
sns.lineplot(Decile_analysis_test.index, Decile_analysis_test.pred,markers=True)
plt.legend(['Actual','Predicted'], loc='best', fontsize=16)
plt.title('Decile Analysis - Total Weekly Sales / Predicted - Test Data', fontsize=18)
plt.ylabel('Total Weekly Sales', fontsize=16)
plt.xlabel('Deciles', fontsize=16)
plt.show()
```
